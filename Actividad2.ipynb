```python
import numpy as np
import pandas as pd
# Columnas numéricas
cols_numericas = [
    "distancia_km", "calorias", "fc_media", "fc_max",
    "cadencia_media", "cadencia_max", "ascenso", "descenso",
    "zancada_m", "punt_estrés_entrenamiento", "pasos", "num_vueltas",
    "alt_min", "alt_max"
]

# Limpieza y conversión
for col in cols_numericas:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", "")     # quita comas
        .str.replace("--", "")    # quita guiones
        .str.strip()              # quita espacios
        .replace("", np.nan)      # convierte vacíos en NaN
        .astype(float)            # convierte a decimal
    )

# Verificacion de datos
print(df.dtypes)
```

```python
import numpy as np
import pandas as pd
## Eliminar columna debido a que no la usare y no hace referencia en mis datos
if "favorito" in df.columns:
    df = df.drop(columns=["favorito"])

##Validacion
df.columns
print(df.head())
```

```python
##Cuantas calorias se queman por kilometro

if {"calorias", "distancia_km"}.issubset(df.columns):
    df["cal_km"] = np.where(df["distancia_km"] > 0, df["calorias"] / df["distancia_km"], np.nan)
print(df.head())
```

```python
##Creacion de Filtros

# Carreras distancia mayor o igual a 5 km
filtro_distancia = df[df["distancia_km"] >= 5]
filtro_distancia.head()
```

```python
# Subconjunto de solo carreras
df_carreras = df[df["tipo"] == "Carrera"]
df_carreras.head()
```

```python
# Subconjunto de alta intensidad (FC media > 160)
df_alta_intensidad = df[df["fc_media"] > 160]
df_alta_intensidad.head()
```

```python
# Subconjunto de baja intensidad (FC media ≤ 140)
df_baja_intensidad = df[df["fc_media"] <= 140]
df_baja_intensidad.head()
```

Aca dejo 3 subcojuntos mostrados, solo las carreras, y la subconjuntos de alta y baja intensidas de ejecucion de carrera.

```python
# Agrupar por tipo y calcular promedios
grupo_tipo = df.groupby("tipo")[["distancia_km", "calorias", "fc_media"]].mean().round(2)
grupo_tipo
```
